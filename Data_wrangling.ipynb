```python
from conllu import parse
import numpy as np
import pandas as pd
```

```python
# parsing the english universal dependencies currentyl in conll format
dev = open('data/en-ud-dev.conllu','r').read()
eng_ud = parse(dev)
```

```python
# function to get the sentence into a string
get_list = lambda x : [i['form'] for i in x]
get_sentence = lambda x : " ".join(get_list(x))
```

```python

sent = eng_ud[33]
sentence = get_sentence (sent)
print(sentence)
```

```python
# get event structure data
distributive = pd.read_csv('event_structure_udewt_v1/event_structure_pred_arg_udewt_v1.tsv', sep='\t')
distributive_train = distributive[distributive.Split == 'train']
distributive_dev = distributive[distributive.Split == 'dev']
distributive_test = distributive[distributive.Split == 'test']
```

```python
distributive_dev.head(50)
```

```python
distributive_sent = distributive_dev.iloc[45]
print(distributive_sent)
```

```python
# Get the Eng UD sentence id from, from Sentence.ID
sent_id = int(distributive_sent['Sentence.ID'].split()[1])-1
pred_id = [int(i)-1 for i in (distributive_sent['Pred.Span'].split('_'))]
arg_id =  [int(i)-1 for i in (distributive_sent['Arg.Span'].split('_'))]

print(sent_id)
print(pred_id)
print(arg_id)
```

```python
# print the sentence from Eng UD
sentence = get_sentence(eng_ud[sent_id])
print(sentence)
```

```python
#print predicate
sent_list = (get_list(eng_ud[sent_id]))
print('Predicate: ', end=' ')
for i in pred_id:
    print(sent_list[i], end=' ')
print()
print('Argument(s): ', end=' ')
for i in arg_id:
    print(sent_list[i], end=' ')
print()
```
